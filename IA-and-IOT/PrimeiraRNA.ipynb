```python
# Importando bibliotecas

import pandas as pd

# Arquitetura Sequencial --> permite a criação de RNA fully conncected
from keras.models import Sequential

# Dense --> permite a criação de camadas totalmente conectadas
from keras.layers import Dense

# Separação em treino e teste
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
dados = pd.read_csv('/content/darts.csv')
dados.head()
```

```python
dados.shape
```

```python
X = dados.drop('competitor', axis=1)

# Target --> multiclasses (4 classes)
y = dados['competitor']
```

```python
y = pd.get_dummies(y)
y.head()
```

```python
# Convertendo y em colunas booleanaspor classes
encoder = OneHotEncoder()
y = encoder.fit_transform(y.reshape(-1, 1))
```

```python
dados['competitor'].unique()
```

```python
# Dados de treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)
#
```

```python
X_treino.shape, X_teste.shape
```

```python
y_treino.shape, y_teste.shape
```

## Criação do modelo de rede neural


```python
modeloRNA = Sequential()

```

```python
# Input Layer --> 2 neuronios (temos 2 features)
modeloRNA.add(Dense(units=128, activation='relu', input_dim=2))

#Hidden Layer 1 --> 128 neuronios (escolhidos arbitrariamente)
modeloRNA.add(Dense(units=128, activation='relu'))

#Hidden Layer 2 --> 32 neuronios (escolhidos arbitrariamente)
modeloRNA.add(Dense(units=32, activation='relu'))

#Output Layer --> 4 neuronios (temos 4 classes)
modeloRNA.add(Dense(units=4, activation='softmax'))
```

```python
modeloRNA.summary()
```

```python
# treinamento da rede neural

# compile (função de custo --> loss, optimizer --> )
modeloRNA.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
modeloRNA.fit(X_treino, y_treino, epochs=50)
```

```python
# Gerando o array com as previsoes do modelo
previsoes = modeloRNA.predict(X_teste)
previsoes

# entrega as previsões num formado de porcentagem
```
